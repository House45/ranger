```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as mno
from sklearn import linear_model
```

```python
df1=pd.read_csv('/content/TotalOutcome.csv')
```

```python
df1.drop(columns=['MOVE','P_SFAT','HYPTMDCODE01','ANTA01'],inplace=True)
```

```python
df1.head()
```

```python
missing_columns = df1.columns
```

```python
def random_imputation(df, feature):

    number_missing = df[feature].isnull().sum()
    observed_values = df.loc[df[feature].notnull(), feature]
    df.loc[df[feature].isnull(), feature] = np.random.choice(observed_values, number_missing, replace = True)
    
    return df
```

```python
df = pd.DataFrame(np.zeros(df1.shape),columns=df1.columns)
for feature in missing_columns:
    df[feature] = df1[feature].copy()
    df = random_imputation(df1, feature)
```

```python
df.head()
```

```python
y = df1.CLASS.values
df1.drop(columns='CLASS',inplace=True)
```

```python
import numpy as np
y = np.array(y).reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,roc_auc_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.33, random_state=42)

clf = LogisticRegression(C=1.0, penalty= 'l2', solver='liblinear',class_weight='balanced').fit(X_train, y_train)
pred_visit_2 = clf.predict(X_test)

accuracy_score(pred_visit_2,y_test)
```

```python
roc_auc_score(y_test,pred_visit_2)
```
